# Prepare your data

```python
#System prompt: Suppose you are doctor. Patients will ask you questions. You will provide treatment based on their symptoms.
```

```python
import pandas as pd
import json

# Assuming df is your DataFrame containing "q_body" and "a_body" columns

def create_json_message(system_content, user_content, assistant_content):
    message = {
        "messages": [
            {"role": "system", "content": system_content},
            {"role": "user", "content": user_content},
            {"role": "assistant", "content": assistant_content}
        ]
    }
    return message

json_list = []

for index, row in df_train.iterrows():
    system_content = "Suppose you are doctor. Patients will ask you questions. You will provide treatment based on their symptoms."
    user_content = row['q_body']
    assistant_content = row['a_body']

    json_message = create_json_message(system_content, user_content, assistant_content)
    json_list.append(json_message)

# Writing the JSON data to separate lines in a file
with open('output1.jsonl', 'w', encoding='utf-8') as jsonl_file:
    for json_message in json_list:
        jsonl_file.write(json.dumps(json_message, ensure_ascii=False) + '\n')
```

```python
input_file = 'output1.jsonl'

data_list = []
with open(input_file, 'r', encoding='utf-8') as jsonl_file:
    for line in jsonl_file:
        data_list.append(json.loads(line))
```
